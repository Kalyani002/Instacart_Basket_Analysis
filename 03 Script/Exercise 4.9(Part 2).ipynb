```python
# Creating a sample.
np.random.seed(4)
```

```python
## Creating a List to hold true/false values of dataframe
dev = np.random.rand(len(instacart)) <= 0.7
```

```python
dev
```

```python
# WE can use your list to split your dataframe into two samples
# Store 70% of sample in dataframe big
big = instacart[dev]
```

```python
# Store 30% of the sample in dataframe small
small = instacart[~dev]
```

```python
# Let's check length of the original dataframe and the newly created sample data frame by using len() function
len(instacart)
```

```python
len(big) + len(small)
```

Since the numbers match, we can assume that your subsetting procedure was at least successful

STEP 7: To visualize relationship between age and income, create a scatterplot using the sns.scatterplot() function.

```python
# Reducing your samples to only those columns necessary for the chart
instacart_4 = small[['age', 'income']]
```

```python
# Check dataframe
instacart_4.head()
```

```python
# Create scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

The scatterplot shows a strong relationship between age and income. As per scatterplot, as age increases the income also increases. 
Particularly the age between 40 to 80, the higher the age the higher the income. The age group 20 to 40 falls under the lower income category.
