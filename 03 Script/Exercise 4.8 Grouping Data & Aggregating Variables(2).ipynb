# Step 2

```python
# find the aggregated mean of the “order_number” column grouped by “department_id” for  entire dataframe.
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

# Step 3 -- Analyze the result
The subset of dataframe had only 8 'drpartment_id', whereas entire dataframe contains 21 'department_ids'

```python
#  A new column called “max_order” is created
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking new column "max_order"
ords_prods_merge.head(30)
```

# Step 4: Deriving Columns with loc()

```python
# Assigning loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#  Access multiple columns at the same time.
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# Step 5: check the basic statistics of the product prices for each loyalty category.

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

-- New customer spends $3 extra compared to loyal customer.

# Step 6:  Create a spending flag for each user based on the average price across all their orders

```python
#  A new column called “mean_of_prices” is created
ords_prods_merge['mean_of_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking new column 'mean_of_prices'
ords_prods_merge.head(20)
```

```python
# Assigning spender flag
ords_prods_merge.loc[ords_prods_merge['mean_of_prices'] < 10, 'spender_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_of_prices'] >= 10, 'spender_flag'] = 'High Spender'
```

```python
# Checking frequency of new column 'spender_flag'.
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
#  Access multiple columns at the same time.
ords_prods_merge[['user_id', 'spender_flag', 'mean_of_prices', 'prices']].head(60)
```

# Step 7:Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
#  A new column called “mean_of_prices” is created
ords_prods_merge['median_of_days'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
# Checking new column 'median_oF_days'
ords_prods_merge['median_of_days'].head(60)
```

```python
# Assigning frequency flag

ords_prods_merge.loc[ords_prods_merge['median_of_days'] > 20, 'frequency_flag'] = 'Non_frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_of_days']) > 10 & (ords_prods_merge['median_of_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_of_days'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
# Checking frequency  of 'frequency_flag' column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#  Access multiple columns at the same time.
ords_prods_merge[['user_id', 'frequency_flag', 'median_of_days', 'days_since_previous_order']].head(60)
```
